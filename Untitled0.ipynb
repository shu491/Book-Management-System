```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# TASK 1: CREATE A DATAFRAME FROM A DICTIONARY
# Define the raw data as a dictionary
data = {
    "Student_Name" : ["Alice Wong", "Maria Alonzo", "Larry Smith", "John Liew", "Raymond Arbough"],
    "Coursework_Mark": [35,40,39,43,48],
    "Final_Exam_Mark": [32,36,31,36,37],
}
```

```python
# Calculate Final Score using weightage: 60% coursework + 40% exam
data["Final_Score"] = [
    int(cw + fe)
    for cw, fe in zip(data["Coursework_Mark"], data["Final_Exam_Mark"])
]
```

```python
# Create the DataFrame
df = pd.DataFrame(data)
# Display the DataFrame
print(df)
```

```python
# TASK 2: APPEND NEW ROWS USING pd.concat()

# Define the new students as a dictionary
new_data = {
    "Student_Name": ["Olivia Nielson", "Charlie Mund"],
    "Coursework_Mark": [28, 27],
    "Final_Exam_Mark": [32, 30],
    "Final_Score": [60, 57],
}
```

```python
# Convert to a DataFrame
new_df = pd.DataFrame(new_data)

# Concatenate the two DataFrames
combined_df = pd.concat([df, new_df], ignore_index=True)

# Display the result
print(combined_df)
```

```python
# TASK 3: ADD A NEW COLUMN
# Define the grades in the same order as the students in combined_df
grades = ["B", "A-", "B+", "A-", "A", "B-", "C+"]

# Add the Grades column to the DataFrame
combined_df["Grades"] = grades

# Display the updated DataFrame
print(combined_df)
```

```python
# TASK 4: DELETE A ROW USING BOOLEAN INDEXING

# Keep only rows where the Student_Name is NOT "Larry Smith"
combined_df = combined_df[combined_df["Student_Name"] != "Larry Smith"]

# Reset the index so row numbers are continuous again
combined_df = combined_df.reset_index(drop=True)

#Confirm the deletion
print(combined_df)
print(f"\nRows remaining: {len(combined_df)}")
```

```python
# TASK 5: BAR CHART AND CSV EXPORT
# Count how many times each grade appears
grade_counts = combined_df["Grades"].value_counts()

# Plot a bar chart
plt.figure(figsize=(8, 5))
plt.bar(grade_counts.index, grade_counts.values, color="steelblue")
plt.xlabel("Grade")
plt.ylabel("Number of Students")
plt.title("Grade Distribution — SWA2124")
plt.xticks(rotation=0)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()

# Save the DataFrame to a CSV file
combined_df.to_csv("swa2124_students.csv", index=False)
print("swa2124_students.csv")
```
